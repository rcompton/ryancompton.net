```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv('hall_effect_log.csv')
print(df.sample(5))

# Convert time to relative seconds for better readability
df['Relative Time'] = df['Time'] - df['Time'].iloc[0]

# Set Seaborn theme for better aesthetics
sns.set_theme(style="whitegrid")

# Create a figure and axes
plt.figure(figsize=(14, 8))

# Plot Sensor Values over Time
sns.lineplot(x="Relative Time", y="Raw Sensor 0", data=df, label="Raw Sensor 0", marker=None, linewidth=1)
sns.lineplot(x="Relative Time", y="Corrected Sensor 0", data=df, label="Corrected Sensor 0", marker=None, linewidth=1)
#sns.lineplot(x="Relative Time", y="Sensor Value 1", data=df, label="Sensor Value 1", marker='.', linewidth=1)
#sns.lineplot(x="Relative Time", y="Sensor Value 2", data=df, label="Sensor Value 2", marker='.', linewidth=1)

# Title and labels
plt.title("Sensor Values Over Time", fontsize=18, fontweight='bold')
plt.xlabel("Time (seconds)", fontsize=14)
plt.ylabel("Sensor Values (V)", fontsize=14)

# Add a legend
plt.legend(title="Sensors", fontsize=12, title_fontsize=14)

# Show grid
plt.grid(True, linestyle="--", alpha=0.6)

# Display the plot
plt.show()

```

```python
plt.figure(figsize=(14, 8))

df['mean12'] = (df['Sensor Value 1'] + df['Sensor Value 2'])/2

# Plot Sensor Values over Time
#sns.lineplot(x="Relative Time", y="Sensor Value 1", data=df, label="Sensor Value 1", marker='.', linewidth=1)
#sns.lineplot(x="Relative Time", y="Sensor Value 2", data=df, label="Sensor Value 2", marker='.', linewidth=1)
sns.lineplot(x="Relative Time", y="mean12", data=df, label="mean12", marker='.', linewidth=1)

# Title and labels
plt.title("Sensor Values Over Time", fontsize=18, fontweight='bold')
plt.xlabel("Time (seconds)", fontsize=14)
plt.ylabel("Sensor Values (V)", fontsize=14)

# Add a legend
plt.legend(title="Sensors", fontsize=12, title_fontsize=14)

# Show grid
plt.grid(True, linestyle="--", alpha=0.6)

# Display the plot
plt.show()
```
