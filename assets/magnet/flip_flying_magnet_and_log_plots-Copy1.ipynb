```python
import numpy as np
import pandas as pd
import csv
import matplotlib.pyplot as plt

data = pd.read_csv("pid_duty_magnet_data.csv")

# Create the plot
plt.figure(figsize=(12, 6))
plt.plot(data["Time (s)"], data["Sensor 0 Voltage"], label="Hall Sensor 0 Voltage")
plt.plot(data["Time (s)"], data["Sensor 1 Voltage"], label="Hall Sensor 1 Voltage")


v_min = min(data["Sensor 0 Voltage"].min(), data["Sensor 1 Voltage"].min())
v_max = max(data["Sensor 0 Voltage"].max(), data["Sensor 1 Voltage"].max())

# Highlight magnet ON/OFF periods
plt.fill_between(data["Time (s)"],
                 v_min, v_max,  # Fill between v_min and v_max vertically
                 where=data["Magnet State"] == 1,
                 color='green', alpha=0.2, label="Magnet ON")  # No need for transform here

# Add labels and title
plt.xlabel("Time (s)")
plt.ylabel("Hall Sensor Voltage (V)")
plt.title("Hall Sensor Voltage with Rise and Fall Times")
plt.legend(loc="upper right")
plt.grid(True)

# Show the plot
plt.show()
```
