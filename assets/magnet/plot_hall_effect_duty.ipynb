```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")



# Load the data
df = pd.read_csv('hall_effect_log.csv')
print(df.sample(5))

# Convert time to relative seconds for better readability
df['Relative Time'] = df['Time'] - df['Time'].iloc[0]
df['Raw Sensor Difference'] = df['Raw Sensor 0'] - df['Raw Sensor 1']


# Create a figure and axes
plt.figure(figsize=(14, 8))

# Plot Sensor Values over Time
sns.lineplot(x="Relative Time", y="Raw Sensor 0", data=df, label="Raw Sensor 0", marker=None, linewidth=1)
sns.lineplot(x="Relative Time", y="Raw Sensor 1", data=df, label="Raw Sensor 1", marker=None, linewidth=1)
sns.lineplot(x="Relative Time", y="Corrected Sensor Difference", data=df, label="Corrected Sensor Difference", marker=None, linewidth=1)
sns.lineplot(x="Relative Time", y="Raw Sensor Difference", data=df, label="Raw Sensor Difference", marker=None, linewidth=1)


# Title and labels
plt.title("Sensor Values Over Time", fontsize=18, fontweight='bold')
plt.xlabel("Time (seconds)", fontsize=14)
plt.ylabel("Sensor Values (V)", fontsize=14)

# Add a legend
plt.legend(title="Sensors", fontsize=12, title_fontsize=14)

# Show grid
plt.grid(True, linestyle="--", alpha=0.6)

# Display the plot
plt.show()

```
