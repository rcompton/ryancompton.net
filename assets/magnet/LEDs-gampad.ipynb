```python
from gpiozero import LED
import time
import evdev
```

```python
gamepad = evdev.InputDevice('/dev/input/event4')
print(gamepad)
```

```python
white = LED(2)
yellow = LED(20)
orange = LED(3)
green = LED(4)
blue = LED(15)
red = LED(21)
```

```python
for event in gamepad.read_loop():
    if event.type == evdev.ecodes.EV_KEY:
        print(evdev.categorize(event))
        keyevent = evdev.categorize(event)
        if keyevent.keystate == evdev.KeyEvent.key_down:
            if keyevent.scancode == 304:
                red.toggle()
            elif keyevent.scancode == 308:
                blue.toggle()
            elif keyevent.scancode == 307:
                green.toggle()
            elif keyevent.scancode == 305:
                orange.toggle()
            elif keyevent.scancode == 311:
                white.toggle()
            elif keyevent.scancode == 313:
                yellow.toggle()
```

```python
for i in range(500):
    blue.on()
    time.sleep(0.01*(i%11))
    blue.off()
    
    green.on()
    time.sleep(0.01*(i%22))
    green.off()

    orange.on()
    time.sleep(0.01*(i%33))
    orange.off()

    yellow.on()
    time.sleep(0.01*(i%33))
    yellow.off()

    white.on()
    time.sleep(0.01*(i%44))
    white.off()
    
    red.on()
    time.sleep(0.01*(i%55))
    red.off()
```
