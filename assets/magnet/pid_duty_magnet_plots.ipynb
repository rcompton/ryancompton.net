```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')

# Load the data from the CSV file
data = pd.read_csv("pid_duty_magnet_data.csv")

# Convert the timestamp to relative time (starting from 0)
data["Time_s"] = data["Time_s"] - data["Time_s"].min()

# Create the figure and the two subplots (one for voltage, one for duty cycle)
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(12, 8))

# --- Plot Hall Sensor Voltage on the top subplot (ax1) ---
ax1.plot(data["Time_s"], data["HallVoltage1"], label="Hall Sensor Voltage")

# Plot Setpoint
ax1.plot(data["Time_s"], data["Setpoint"], label="Setpoint", color="green", linestyle="-.")

# Highlight Hysteresis Levels
ax1.plot(data["Time_s"], data["HYST_LOW"], label="Hysteresis Low", color="purple", linestyle=":")
ax1.plot(data["Time_s"], data["HYST_HIGH"], label="Hysteresis High", color="orange", linestyle=":")

# Add labels and title for ax1
ax1.set_ylabel("Voltage (V)")
ax1.set_title("Hall Sensor Voltage, Duty Cycle, Setpoint, and Hysteresis")
ax1.legend()
ax1.grid(True)

# --- Plot Duty Cycle on the bottom subplot (ax2) ---
ax2.plot(data["Time_s"], data["DutyCycle"], label="Duty Cycle", color="red")

# Add labels for ax2
ax2.set_xlabel("Time (s)")
ax2.set_ylabel("Duty Cycle")
ax2.legend()
ax2.grid(True)

# Align the x-axes
plt.setp(ax1.get_xticklabels(), visible=False)  # Hide x-ticks for the top plot
plt.subplots_adjust(hspace=.01)  # Remove vertical space between subplots

# Show the plot
plt.show()
```
