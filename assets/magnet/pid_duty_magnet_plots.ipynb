```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV data
file_path = "hybrid_bangbang_pid_log.csv"  # Update with the correct path if necessary
data = pd.read_csv(file_path)
print(data.sample(5))

# Convert time to float for proper plotting
data["Time_s"] = data["Time_s"].astype(float)

# Plot raw sensor values over time
plt.figure(figsize=(12, 6))
plt.plot(data["Time_s"], data["Sensor0_Raw"], label="Sensor0 Raw Voltage", linestyle="-")
plt.plot(data["Time_s"], data["Sensor1_Raw"], label="Sensor1 Raw Voltage", linestyle="-")
plt.plot(data["Time_s"], data["HYST_HIGH"], label="HYST_HIGH", linestyle="--")
plt.plot(data["Time_s"], data["HYST_LOW"], label="HYST_LOW", linestyle="--")


plt.title("Sensor Voltage Over Time", fontsize=16)
plt.xlabel("Time (s)", fontsize=14)
plt.ylabel("Voltage (V)", fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)
plt.tight_layout()

# Plot the duty cycle over time
plt.figure(figsize=(12, 6))
plt.plot(data["Time_s"], data["DutyCycle_%"], label="Duty Cycle (%)", color="orange")

plt.title("Duty Cycle Over Time", fontsize=16)
plt.xlabel("Time (s)", fontsize=14)
plt.ylabel("Duty Cycle (%)", fontsize=14)
plt.grid(True)
plt.legend(fontsize=12)


plt.figure(figsize=(12, 6))
plt.plot(data["Time_s"], data["Sensor0-Sensor1"], label="Sensor0-Sensor1", linestyle="-")
plt.legend(fontsize=12)

plt.show()

```
