```python
import requests
import re
from bs4 import BeautifulSoup
import json
from collections import deque
```

```python
def get_nearby(city):
    url = 'https://www.padmapper.com/apartments/'+city
    print(url)
    r = requests.get(url,
                 headers = 
        {"accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7",
        "accept-language": "en-US,en;q=0.9",
        "sec-ch-ua": "\"Not/A)Brand\";v=\"99\", \"Google Chrome\";v=\"115\", \"Chromium\";v=\"115\"",
        "sec-ch-ua-mobile": "?0",
        "sec-ch-ua-platform": "\"macOS\"",
        "sec-fetch-dest": "document",
        "sec-fetch-mode": "navigate",
        "sec-fetch-site": "none",
        "sec-fetch-user": "?1",
        "upgrade-insecure-requests": "1",
        "cookie": "padmapper.sid=s%3AvVYONpVDCtkHRRTGNgShqYAWqRkwh4sT.DKzLzKB0hlJCYSnfLJuOeRCfP3tK7VfYyDWVO8Y2oX8; _uc_referrer=https://www.google.com/; _pbjs_userid_consent_data=3524755945110770; csrftoken=JeHNj5LLo0me4EwVSrVPD3juq8RSRuKRRhzLFfORjUXJPq1cOOWxnowXwJZl5o2Vk5VJarAwo4sqT34L"
        })
    
    # Load the HTML document
    soup = BeautifulSoup(r.content, 'html.parser')
    # Get all script tags
    script_tags = soup.find_all('script')
    output = set()
    for script_tag in script_tags:
        if "window.__PRELOADED_STATE__" in script_tag.text:
            try:
                jj = json.loads(re.search(r"window.__PRELOADED_STATE__ = (.*?);", script_tag.text).group(1))
                for city in jj['currentSearch']['city']['data']['nearby']['cities']:
                    output.add(city['path'])
            except:
                print(script_tag)
    return output
    

get_nearby('burlingame-ca')
```

```python
visited = set()
to_visit = deque()
to_visit.append('burlingame-ca')

def process_curr_node(curr):
    visited.add(curr)
    nearbys = get_nearby(curr)
    for nearby in nearbys:
        if nearby not in visited:
            to_visit.append(nearby)
    
while len(visited) < 35:
    curr = to_visit.popleft()
    process_curr_node(curr)
    for city in sorted(visited):
        print("visited: ", city)
```
